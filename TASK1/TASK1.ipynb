```python
# Outlier Detection
plt.figure(figsize=(8,5))
sns.boxplot(x=df['MonthlyIncome'])
plt.title('Monthly Income Outliers')
plt.show()
```

```python
# EDA Visualizations
# Age Distribution
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# Attrition count
sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition')
plt.show()
```

```python
# Attrition by Department
sns.countplot(x='Department', hue='Attrition', data=df)
plt.xticks(rotation=20)
plt.show()
```

```python
# Overtime vs Attrition
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.show()
```

```python
# Monthly Income vs Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.show()
```

```python
# Job Satisfaction vs Attrition
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(15,10))
sns.heatmap(df.select_dtypes(include=np.number).corr(),
            cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df, palette='Set2')
plt.title('Employee Attrition Count')
plt.show()
```

```python
sns.scatterplot(
    x='Age',
    y='MonthlyIncome',
    hue='Attrition',
    data=df,
    palette='viridis'
)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Department', hue='Attrition', data=df,palette = 'Set3')
plt.title('Department-wise Attrition')
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df,palette='viridis')
plt.title('Overtime vs Attrition')
plt.show()
```

```python
attrition_counts = df['Attrition'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    attrition_counts,
    labels=attrition_counts.index,
    autopct='%1.1f%%'
)
plt.title('Attrition Percentage')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(
    x='Attrition',
    y='JobSatisfaction',
    data=df,
    palette = 'rocket'
)
plt.title('Job Satisfaction vs Attrition')
plt.show()
```

**INSIGHTS**


Employees working overtime tend to have higher attrition.


Certain departments experience higher employee turnover.


Lower monthly income groups show relatively higher attrition.


Younger employees tend to leave more frequently.


Employees with fewer years at the company have higher attrition rates.
